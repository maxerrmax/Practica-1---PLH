```python
import re

def preprocess_lines(text):
    """
    Processa el text línia per línia:
    - Elimina dígits
    - Posa en minúscules
    - Normalitza espais
    - Uneix totes les línies amb doble espai
    - Afegeix doble espai al principi i final
    """
    processed_lines = []

    for line in text.splitlines():
        # 1️⃣ Eliminar dígits
        line = re.sub(r"\d", "", line)
        
        # 2️⃣ Minúscules
        line = line.lower()
        
        # 3️⃣ Normalitzar espais
        line = re.sub(r"\s+", " ", line).strip()
        
        if line:  # Evitar línies buides
            processed_lines.append(line)
    
    # 4️⃣ Unir totes les línies amb doble espai
    result = "  ".join(processed_lines)
    
    # 5️⃣ Afegir doble espai al principi i final
    result = "  " + result + "  "
    
    return result


if __name__ == "__main__":
    input_file = "langId/spa_tst.txt"
    with open(input_file, "r", encoding="utf-8") as f:
        text = f.read()
    
    processed_text = preprocess_lines(text)
    print(processed_text[:5000])  # Mostra els primers 5000 caràcters
```
