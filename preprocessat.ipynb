```python
import re

def preprocess(text):
    # 1️⃣ Eliminar dígits
    text = re.sub(r"\d", "", text)
    
    # 2️⃣ Minúscules
    text = text.lower()
    
    # 3️⃣ Normalitzar espais
    text = re.sub(r"\s+", " ", text).strip()
    
    # 4️⃣ Separar frases només per puntuació universal
    sentences = re.split(r'(?<=[.!?])\s+', text)
    
    # 5️⃣ Unir amb doble espai
    text = "  ".join(sentences)

    # 6️⃣ Afegir doble espai al principi i final
    text = "  " + text + "  "
    
    return text


if __name__ == "__main__":
    input_file = "langId/spa_trn.txt"  # Ajusta la ruta si cal
    
    with open(input_file, "r", encoding="utf-8") as f:
        text = f.read()
    
    processed_text = preprocess(text)
    
    print(processed_text[:5000])  # Mostra els primers 500 caràcters del text processat
```
